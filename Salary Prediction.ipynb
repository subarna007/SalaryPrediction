```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Salary Data.csv')
```

```python
dataset = dataset.dropna()     #drops rows having empty value or NAN value
dataset
```

```python
dataset = dataset[['Years of Experience','Salary']]
dataset
```

```python
dataset.iloc[:,0:1].values
```

```python
X = dataset.iloc[:, 0:-1].values
X
```

```python
Y = dataset.iloc[:,1:].values
Y
```

```python
# train size-> min 2/3
# test size -> 1/3

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1,random_state = 1)
```

```python
print(X_train)
```

```python
total_len = (len(X))
print("All total length = ",total_len)
print("X_train length = ", len(X_train))
print("X_test length = ", len(X_test))
```

```python
#Training model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
#testing model
exp = float(input("Enter years of experience = "))
exp = np.array([[exp]])
print(exp)

# # # # predict the salary
predicted_salary = model.predict(exp)
# print(predicted_salary)
print("Your predicted_salary = ",predicted_salary)
```
